### **Write a query that returns tag\_id, the number of times each tag\_id is used and the tag\_name.**  

### Use the GROUP BY and HAVING clause to narrow your focus.

Try multiple keywords, ie woman vs female (female ended up giving better results)

```sql
-- Solution
select COUNT(tag_id) as count_id,  goodreads_book_id
from BooksDB.dbo.book_tags
GROUP BY goodreads_book_id
```

```sql
USE BooksDB
SELECT book_tags.tag_id AS 'TAG ID' ,tags.tag_name AS 'TAG NAME',
SUM(book_tags.count) AS 'TAG COUNT'
FROM book_tags
INNER JOIN tags
ON book_tags.tag_id=tags.tag_id
GROUP by book_tags.tag_id, tags.tag_name 
HAVING tag_name LIKE '%WOMAN%'
ORDER BY SUM(book_tags.count) DESC;
```

### **Write a query to return authors, titles, ratings and** **selected tag(s)\_id.**

```sql
-- Solution
USE BooksDB
SELECT book_tags.tag_id, books.authors, books.title, books.average_rating
FROM book_tags
INNER JOIN books
ON book_tags.goodreads_book_id = books.book_id
WHERE book_tags.tag_id = 11643


ORDER BY books.authors, books.title, books.average_rating DESC
```

```sql
-- Event 1 Query
USE BooksDB
SELECT book_tags.tag_id AS 'TAG ID' ,tags.tag_name AS 'TAG NAME',
SUM(book_tags.count) AS 'TAG COUNT'
FROM book_tags
INNER JOIN tags
ON book_tags.tag_id=tags.tag_id
GROUP by book_tags.tag_id, tags.tag_name 
HAVING tag_name LIKE '%Father%' or tag_name LIKE '%Dad%'
ORDER BY SUM(book_tags.count) DESC;
```

```sql
SELECT books.title, books.authors, COUNT(tr.book_id) AS number_saves, tags.tag_id
FROM BooksDB.dbo.books AS books 
INNER JOIN BooksDB.dbo.book_tags AS tags 
ON books.best_book_id = tags.goodreads_book_id
INNER JOIN BooksDB.dbo.to_read AS tr 
ON tags.goodreads_book_id = tr.book_id
WHERE tags.tag_id = 27919 OR tags.tag_id = 4708 OR tags.tag_id = 4710
GROUP BY books.title, books.authors, tags.tag_id
ORDER BY number_saves DESC 
```
